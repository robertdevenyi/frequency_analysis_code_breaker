```python
# delete some 1-2 letter words
stayers = ["a", "A", "I","i","as", "am", "an", "at", "by","be","do", "if", "in","is","it", "my","me","no", "of", "oh", "on", "or", "so", "to", "us", "we", "he", "hi"]

file = open("usa.txt", "r")
outfile = open("USA.txt", "w")
for line in file.readlines():
    if len(line.rstrip()) in [1,2] and line.rstrip() not in stayers:
        pass
    else: 
        outfile.write(line)


#
```
